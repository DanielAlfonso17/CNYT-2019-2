```python

# El formato de cada entrada de la lista es: [Nombre_Ciudad, Latitud, Longitud]

Ciudades = [
    ['Londres', 51.5, -0.1167],
    ['BogotÃ¡', 4.5964, -74.0833],
    ['Roma', 41.8960, 12.4833],
    ['Melbourne', -37.8200, 144.9750],
    ['Nairobi', -1.2833, 36.8167],
    ['Santiago', -33.4500, -70.6670],
    ['Nueva York', 40.6943, -73.9249],
    ['Los Angeles', 34.1139, -118.4068],
    ['MoscÃº', 55.7522, 37.6155],
    ['Quito', -0.2150, -78.5001],
    ['Tokio', 35.6850, 139.7514],
    ['Lagos', 6.4433, 3.3915],
    ['Atenas', 37.9833, 23.7333],
    ['BelÃ©n de ParÃ¡', -1.4500, -48.4800],
    ['Budapest', 47.5000, 19.0833],
    ['Vancouver', 49.2734, -123.1216],
    ['ParÃ­s', 48.8667, 2.3333],
    ['Puebla', 19.0500, -98.2000],
    ['Sidney',-33.854816,151.216454],
    ['Luanda',-8.82727,13.2439512],
    ['Artashat',39.88651,44.579421],
    ['Copenhague',55.6867243,12.5700724],
    ['HongKong',22.349216,114.1857978]
]

print("Las ciudades agregadas son: Sidney, Luanda, Artashat,Copenhague,HongKong")
def gradosRadianes(grados):
    radianes = (math.pi*grados)/180
    return radianes 

def geograficaMatematica(geograficos):
    pendiente = (0-180)/(90-(-90))
    ecuacion = (pendiente*geograficos)+90
    return ecuacion 
    
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

phi, theta = np.mgrid[0 : 2 * np.pi : 0.1, 0 : np.pi : 0.1]  # Equivale a hacer un doble for

# Para pasar de coordenadas esfÃ©ricas (phi, theta) a coordenadas cartesianas (x, y, z):

x = np.cos(phi) * np.sin(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(theta)
ax.plot_wireframe(x, y, z, color='black', rstride=2, cstride=2, linewidth=0.5, alpha=0.2)
for i in Ciudades:
    latitud=i[1]
    longitud=i[2]
    
    latitudMatematico = geograficaMatematica(latitud)
    xPunto =  np.cos(gradosRadianes(longitud)) * np.sin(gradosRadianes(latitudMatematico))
    yPunto = np.sin(gradosRadianes(longitud)) * np.sin(gradosRadianes(latitudMatematico))
    zPunto = np.cos(gradosRadianes(latitudMatematico))
    ax.scatter(xPunto, yPunto, zPunto, color='red', s=10) # El valor de s modifica el grosor del punto
    ax.text(xPunto, yPunto, zPunto, i[0], color='black')






# Para graficar la esfera (estructura de alambre):



# Para graficar un punto cualquiera en el espacio 3d, en este caso (2, -2, 1):



# Para hacer una etiqueta, en este caso sobre el punto (2, -2, 1):



# Para graficar una flecha, en este caso del origen al punto (2, -2, 1):





# Nombres de los ejes

ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')

plt.show()
```

```python
kets = [
    ['|Up>', 0, 1],
    ['|Down>', 90, 0],
    ['|In>', 45, 90],
    ['|Out>', 45, -90],
    ['|+>', 45, 0],
    ['|->', 45, 180]
    
    
]


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

phi, theta = np.mgrid[0 : 2 * np.pi : 0.1, 0 : np.pi : 0.1] 



x = np.cos(phi) * np.sin(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(theta)



ax.plot_wireframe(x, y, z, color='gray', rstride=2, cstride=2, linewidth=0.5, alpha=0.8)




for i in range(len(kets)):

    

    phi= math.radians(kets[i][2])
    theta= 2*math.radians(kets[i][1])

    x = np.cos(phi) * np.sin(theta)
    y = np.sin(phi) * np.sin(theta)
    z = np.cos(theta)


    ax.scatter(x, y, z, color='red', s=15) # El valor de s modifica el grosor del punto

    

    ax.text(x, y, z, kets[i][0], color='black')

    

    ax.quiver(0, 0, 0, x, y, z, color='blue')

    # Nombres de los ejes

    ax.set_xlabel('Eje X')
    ax.set_ylabel('Eje Y')
    ax.set_zlabel('Eje Z')

    plt.show()
```
